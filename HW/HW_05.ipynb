# 1
Apples in new york are expensive, is just a statement until it can be testable. It is testable when we have defined somthing it can be tested against. For example, an apple costing $1.75 is expensive. Now Apples in new york becomes a hypothesis. Data that is vague or subjective can not be tested / examined statistically, for example, who is going to be a better Prime Minister Jagmeet Singh or Justin Trudeau, we don't know because Jagmeet is / was not Prime Minister, so there is no data on him. The data should be able to be categorized or quantified. A good null hypothesis should be testable, be able to be proven false and precisely defined. The Null hypothesis is an idea that can be tested, the data proposes no change from your idea. An alternative hypothesis, rejects the null hypothesis and propses that there is a real effect or difference in the data than expected. You're aiming to prove the alternative hypothesis. 

# 2
A population is the group we are trying to study. A sample, is a portion of the population which we use to represent the population. It impractical to gather data from every single person in the population. A parameter is the measurement of the population (often we don't know the actualy parameter, but we can create a hypothetical population based on the sample). Gamme distributions have a shape paramter... Normal distributions have a mean parameter... A statistic is a measurement of the sample, for example, the sample mean. 
mu: population mean
x bar: sample mean
mu_0: hypothesize population mean (not in the text book apparently - notebookLM) 
x_i: A single observation within a sample. used to calculate the sample statistic, you use it when you do statistical calculation by hand. and the programs use it to. x_1 height of the first person. x_2, height of the second.

The tests refer to the population parameter rather than the sample statistic means that the test aims to provide conclusion about the population and not the sample. The only reason we are testing the sample is to propose something about the population. 

Example from the textbook: You're baking a cake and want to check if it's fully cooked. You wouldn't eat the entire cake to find out. You'd insert a toothpick into the center—that's your sample. If it comes out clean, you infer the whole cake—the population—is baked. You use information from the sample (toothpick) to judge the population (cake).

# 4
a smaller p-value makes the null hypothesis look more ridiculous. Say you have a sample, and from that sample you create a hypothetical distribution to represent the population. Now, you create a new hypothetical distribution of how you're data would look if the null hypothesis was used as one of the parameters. Now, overlap the two distributions (one red, and one blue, so you can see the difference of course). If the two distributions overlap eachother alot, the p-value is higher because there is a bigger percentage of the distributions overlapping. This means that your data supports the null hypothesis that you created, the null hypothesis, an idea that you assume about the data. Now, if the p-value was really low, that means the distributions don't overlap eachother very much. If they don't overlap eachother, then you're null hypothesis looks way different from the sample and your null hypothesis can be considered ridiculous.

The population is every student in STA130.

The sample is 80 STA130 students. 

Parameter of interest (p): The proportion of students who can correctly identify if the milk or tea was poured first. Reflects the populations ability to correctly predict. 

Observerd Test Statistic (p tilde): The amount of students who predicted correctly. 49/80


Formal Null Hypothesis
Null Hypothesis: p = 0.5
Alternative Hypothesis: p~ != 0.5

Null Hypothesis (H_0): The proportion of correct predictions is 0.5
Alternative Hypothesis (H_a): The proportion of correct predictions is > 0.5 


Common Significant Level (Alpha) is 0.05, commonly used threshold in hypothesis testing.

We can reject the null hypothesis based on the p-value calulated below. If the p-value is less than or equal to 0.05, the null hypothesis can be rejected. Our data distribution does not overlap the null hypothesis distribution therefore our hypothesis was way off.

Fishers p-value = 0.0047

Sta130 p-value = 0.0442

Both Null Hypothesis can be rejected by calculating the p-value. 

Consider trying to reject the null hypothesis using a confidence interval as well to further support your evidence as a p-value is just a peice of the puzzle and does not relay the potential underlying information.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Creating a simple DataFrame for the experiment data
fishers_data = {'Total Tests': [8], 
                 'Correct Predictions': [8], 
                 'Incorrect Predictions': [8 - 8]}

Fishers_df = pd.DataFrame(fishers_data)

# Data for actual predictions
total_tests = Fishers_df['Total Tests'][0]
correct_predictions = Fishers_df['Correct Predictions'][0]

# Null Hypothesis parameters
p_null = 0.5
mean_null = p_null * total_tests  # Expected mean
std_dev_null = np.sqrt(total_tests * p_null * (1 - p_null))  # Expected standard deviation

# Calculate the z-score for the observed value
z_score = (correct_predictions - mean_null) / std_dev_null

# Calculate the p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

# Print the p-value
print(f"P-value: {p_value:.4f}")

# Estimate mean and standard deviation for a hypothetical larger sample
proportion_correct = correct_predictions / total_tests
mean = proportion_correct * total_tests  # Should be 8.0 (total tests)
std_dev = 1  # Arbitrary standard deviation for visualization

# Generate normal distribution data for actual predictions
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 1000)
y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Generate normal distribution data under the null hypothesis
x_null = np.linspace(mean_null - 3 * std_dev_null, mean_null + 3 * std_dev_null, 1000)
y_null = (1 / (std_dev_null * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_null - mean_null) / std_dev_null) ** 2)

# Plotting both distributions
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Simulated Normal Distribution (Actual Data)', color='blue')
plt.plot(x_null, y_null, label='Null Hypothesis Normal Distribution (p = 0.5)', color='red')
plt.title('Overlapping Normal Distributions')
plt.xlabel('Number of Correct Predictions')
plt.ylabel('Probability Density')
plt.axvline(correct_predictions, color='green', linestyle='dashed', linewidth=1, label='Actual Correct Predictions (8)')
plt.axvline(mean_null, color='purple', linestyle='dashed', linewidth=1, label='Expected Mean (4)')
plt.legend()
plt.grid()
plt.show(renderer='png')
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# Creating a simple DataFrame for the experiment data
fishers_data = {'Total Tests': [8], 
                 'Correct Predictions': [8], 
                 'Incorrect Predictions': [8 - 8]}

Fishers_df = pd.DataFrame(fishers_data)

# Given data from the DataFrame
total_tests = Fishers_df['Total Tests'][0]
correct_predictions = Fishers_df['Correct Predictions'][0]
p_null = 0.5  # Null hypothesis proportion

# Sample proportion
p_hat = correct_predictions / total_tests  # Observed proportion

# Calculate the z-score
z = (p_hat - p_null) / np.sqrt((p_null * (1 - p_null)) / total_tests)

# Calculate the p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# Output the results
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")
```
